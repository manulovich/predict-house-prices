## 1 Подготовка и анализ данных

### 1.0 Очистка данных

__План мойки__:
+ Удалим признаки с большим количеством пропусков (Alley, FireplaceQu, PoolQC, Fence, MiscFeature);
+ Удалилим признаки, которые дублируют друг друга (GarageYrBlt)*;
+ Пропуски в категориальных признаках заполним значением 'Other';
+ Пропуски в численных признаках заполним средним значением.

___
*Удалим признак GarageYrBlt, так как он часто сопадает значениям признака YearBuilt

P.S. Все изменение датасета вынесем в функцию, чтобы в дальнейшем их можно было быстро применять для тестовых данных.

```python
def data_washing(df: pd.DataFrame):
    """
    Функция для очистки данных
    """
    df = df.copy()
    
    features_with_a_lot_of_gaps = ['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
    features_duplicate = ['GarageYrBlt']
    
    df = df.drop(features_with_a_lot_of_gaps + features_duplicate, axis=1)
    
    numerical_features, categorical_features = split_numerical_categorical_features(df)
    
    df[categorical_features] = df[categorical_features].fillna('Other')
    df[numerical_features] = df[numerical_features].apply(lambda s: s.fillna(s.mean()))
    
    return df
```

```python
df_with_solution = data_washing(df_with_solution)
```

## 2 Выбор алгоритма и его гиперпараметров

### 2.1 Выбор алгоритма

В качестве алгоритма для предсказания стоимости недвижимости попробуем линейную регрессию и Bagging.
Линейная регрессия была выбрана из-за того, что ее легко интепретировать и просто реализовывать, так как не трубет настройку гиперпараметров.

### 2.2 Подбор гиперпараметров

Гиперпараметры выбрали на этапе обучения.

### 2.3 Подготовка данных для обучения

__План__
+ Закодировать категориальные признаки;
+ Провести нормализацию данных.*

P.S. Нормализация - масштабирование значений до диапазона от [0, 1]. Стандартизация - транфмормация данных до матожидания 0 и стандартного отклонения 1. (Для себя)

*Делаем нормализацию, а не стандартизацию для разнообразния.
___
P.S.S. Вынесем все действия по подготовке данных к обучению в отдельную функцию, дабы в будуем быстро применять все изменения для тестовой выборки

```python
def preparing_data_for_learning(df, one: OneHotEncoder = None):
    """
    Функция для подготовки данных к обучению
    """
    
    _, categorical_features = split_numerical_categorical_features(df)
    df = df.copy()
    
    df_encoder, one, col = OHE(df, categorical_features, one)
    df_encoder_normalize = normalize_df(df_encoder)
    
    return df_encoder_normalize, one
```

```python
X_train_prepar, one = preparing_data_for_learning(X_train)
X_valid_prepar, one = preparing_data_for_learning(X_valid, one)
```

## 3 Обучение модели

### 3.0 Подбор гиперпараметров

Качество модели сильно зависит от количества моделей в BaggingRegressor. Для этого подберем знчения этого параметра, ориентируясь на метрики R2, MAE, MSE

```python
grid_bagg = RandomizedSearchCV(
            BaggingRegressor(LinearRegression()),
            param_distributions = {
                'n_estimators': range(1, 100, 1)
            },
            scoring = 'r2',
            cv = 3,
            n_jobs = -1,
            return_train_score = True,
            n_iter = 10
        )

# Обучим сетку на обучающем датасете при помощи метода fit
grid_bagg.fit(X_train_prepar, y_train)
```

```python
print(f"Лучшая метрика:   {grid_bagg.best_score_}")
print(f"Лучшие параметры: {grid_bagg.best_params_}")
```

```python
grid_grad = RandomizedSearchCV(
            GradientBoostingRegressor(),
            param_distributions = {
                'learning_rate': np.arange(0.1, 1, 0.1, dtype=float),
                'n_estimators': np.arange(1, 300, 1, dtype=int)
            },
            scoring = 'neg_mean_squared_error',
            cv = 3,
            n_jobs = -1,
            return_train_score = True,
            n_iter = 10
        )

# Обучим сетку на обучающем датасете при помощи метода fit
grid_grad.fit(X_train_prepar, y_train)
```

```python
print(f"Лучшая метрика:   {grid_grad.best_score_}")
print(f"Лучшие параметры: {grid_grad.best_params_}")
```

```python
grid_bagg_grad = RandomizedSearchCV(
            BaggingRegressor(GradientBoostingRegressor(**grid_grad.best_params_)),
            param_distributions = {
                'n_estimators': range(1, 200, 1),
                'max_features': np.arange(0.1, 1, 0.1, dtype=float),
            },
            scoring = 'neg_mean_squared_error',
            cv = 3,
            n_jobs = -1,
            return_train_score = True,
            n_iter = 10
        )

# Обучим сетку на обучающем датасете при помощи метода fit
grid_bagg_grad.fit(X_train_prepar, y_train)
```

```python
print(f"Лучшая метрика:   {grid_bagg_grad.best_score_}")
print(f"Лучшие параметры: {grid_bagg_grad.best_params_}")
```

### 3.1 Обучение

```python
%%time

# Обычная линейная регрессия
model_lin_reg = LinearRegression()
model_lin_reg.fit(X_train_prepar, y_train)
y_pred_lin_reg = model_lin_reg.predict(X_valid_prepar)
```

```python
%%time

# Бэггинг
model_bagging = BaggingRegressor(LinearRegression(), **grid_bagg.best_params_)
model_bagging.fit(X_train_prepar, y_train)
y_pred_bagg = model_bagging.predict(X_valid_prepar)
```

```python
%%time

# Регрессия
model_grad = GradientBoostingRegressor(**grid_grad.best_params_)
model_grad.fit(X_train_prepar, y_train)
y_pred_grad = model_grad.predict(X_valid_prepar)
```

```python
%%time

# Бэггинг + Регрессия
model_bagging_grad = BaggingRegressor(GradientBoostingRegressor(**grid_grad.best_params_), **grid_bagg_grad.best_params_)
model_bagging_grad.fit(X_train_prepar, y_train)
y_pred_bagg_grad = model_bagging_grad.predict(X_valid_prepar)
```

```python
%%time

# Стэккинг
estimators = [
    ('LinReg', LinearRegression()),
    ('GradReg', GradientBoostingRegressor(**grid_grad.best_params_)),
    ('KnnReg', KNeighborsRegressor())
]

model_stacking = StackingRegressor(estimators, cv=5)

model_stacking.fit(X_train_prepar, y_train)
y_pred_stack = model_stacking.predict(X_valid_prepar)
```

### 3.2 Проверка метрик

```python
print("Линейная регрессия:")
print('R2:', r2_score(y_valid, y_pred_lin_reg))
print('MAE:', mean_absolute_error(y_valid, y_pred_lin_reg))
print('MSE:', mean_squared_error(y_valid, y_pred_lin_reg))
print('====')
print("Бэггинг:")
print('R2:', r2_score(y_valid, y_pred_bagg))
print('MAE:', mean_absolute_error(y_valid, y_pred_bagg))
print('MSE:', mean_squared_error(y_valid, y_pred_bagg))
print('====')
print("Градиентный бустинг:")
print('R2:', r2_score(y_valid, y_pred_grad))
print('MAE:', mean_absolute_error(y_valid, y_pred_grad))
print('MSE:', mean_squared_error(y_valid, y_pred_grad))
print('====')
print("Бэггинг + Градиентный бустинг:")
print('R2:', r2_score(y_valid, y_pred_bagg_grad))
print('MAE:', mean_absolute_error(y_valid, y_pred_bagg_grad))
print('MSE:', mean_squared_error(y_valid, y_pred_bagg_grad))
print('====')
print("Стэкинг:")
print('R2:', r2_score(y_valid, y_pred_stack))
print('MAE:', mean_absolute_error(y_valid, y_pred_stack))
print('MSE:', mean_squared_error(y_valid, y_pred_stack))
```
